```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

dataset_location = "dataset" # change this if needed, it is assumed that all the csv/pickle files are inside that folder.
```

```python
%%time
df = pd.read_pickle(os.path.join(dataset_location, "yelp_review.pickle"))
```

```python
df
```

```python
print("""Hypothesis: Reviews which have useful-funny-cool different than 0 are more likely to have lower star rating.
Null hypothesis: Average rating for reviews with no useful-funny-cool is %.2f""" % df[(df["useful"] == 0) & (df["funny"] == 0) & (df["cool"] == 0)]['stars'].mean()) 
```

```python
df[(df["useful"] != 0) | (df["funny"] != 0) | (df["cool"] != 0)]['stars'].mean()
```

Welch's T test

```python
stats.ttest_ind(list(df[(df["useful"] == 0) & (df["funny"] == 0) & (df["cool"] == 0)]['stars']), list(df[(df["useful"] != 0) | (df["funny"] != 0) | (df["cool"] != 0)]['stars']), equal_var=False)
```

P value is 0 so we can accept this hypothesis

#### Second Hypothesis

Hypothesis: Businesses in Las Vegas have higher ratings compared to average.<br>
Null Hypothesis: Businesses in Las Vegas have lower ratings compared to average.

```python
%%time
df = pd.read_csv(os.path.join(dataset_location, "yelp_business.csv"))
```

```python
df
```

```python
df[df['city'] =='Las Vegas']
```

We again use Welch's T test because our data sizes are quite different

```python
stats.ttest_ind(df[df['city'] == 'Las Vegas']['stars'], df['stars'] , equal_var=False)
```

```python
print('Average rating in Las Vegas:\t%.2f\nAverage rating for all data:\t%.2f' % (df[df['city'] == 'Las Vegas']['stars'].mean(), df['stars'].mean()))
```

We reject our hypothesis(accept null hypothesis), because p value is too large.
