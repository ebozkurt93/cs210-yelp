From all csv files 2 of them are large enough such that reading those take several minutes. Instead of this we can dump dataframes of those csv files into pickle files, and read from there afterwards.

```python
import pandas as pd
import os
```

```python
dataset_location = "dataset" # change this if needed, it is assumed that all the csv files are inside that folder.
```

```python
%%time
df = pd.read_csv(os.path.join(dataset_location, "yelp_user.csv"))
```

```python
%%time
df.to_pickle(os.path.join(dataset_location, "yelp_user.pickle"))
```

```python
del df
```

```python
%%time
df = pd.read_csv(os.path.join(dataset_location, "yelp_review.csv"))
```

```python
%%time
df.to_pickle(os.path.join(dataset_location, "yelp_review.pickle"))
```

```python
del df
```

```python
%%time
df = pd.read_pickle(os.path.join(dataset_location, "yelp_review.pickle"))
```
